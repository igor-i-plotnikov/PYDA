Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

cmds = {
    'p': 'команда, которая спросит номер документа и выведет имя человека, которому он принадлежит',
    's': 'команда, которая спросит номер документа и выведет номер полки, на которой он находится',
    'l': 'команда, которая выведет список всех документов в формате',
    'as': 'команда, которая спросит номер новой полки и добавит ее в перечень',
    'd': 'команда, которая спросит номер документа и удалит его из каталога и из перечня полок',
    'm': 'команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую',
    'a': 'команда, которая добавит новый документ в каталог и в перечень полок',
    'q': 'выйти'
    
}  

def get_shelf_for_doc(doc=None):
    doc_shelf = None
    if doc:
        doc_number = doc 
    else: doc_number = input('Введите номер документа: ') 
    for shelf, docs_list in directories.items():
        if doc_number in docs_list:
            doc_shelf = shelf           
    if not doc_shelf: print('Ошибка! Указанного номера нет на полках.')
    return doc_shelf

def get_pos_for_doc(doc=None):
    doc_pos = None
    if doc:
        doc_number = doc 
    else: doc_number = input('Введите номер документа: ') 
    for pos, doc in enumerate(documents):
        if doc['number'] == doc_number:
            doc_pos = pos
    if not doc_pos: print('Ошибка! Указанного номера нет в документах.')
    return doc_pos        
    
def get_shelf():
    while True:
        shelf = input('Введите номер целевой полки: ')
        if shelf in directories:
            break
        else:
            is_new_shelf = input('Указанная полка не существует. Создать новую полку? (y/n): ')
            if is_new_shelf == 'y':
                add_shelf(shelf)
                break
            elif is_new_shelf == 'n':
                continue
            else:
                print('Неизвестная команда!')
                return   
    return shelf    
    
def move_doc(doc=None):    
    if doc:
        doc_number = doc 
    else: doc_number = input('Введите номер документа: ')
    if not doc_number: return
    shelf = get_shelf_for_doc(doc_number)
    if not shelf: return
    shelf_n = get_shelf()          
    if shelf_n == shelf[0]:
        print('Документ уже на целевой полке.')
        return 
    directories[shelf].remove(doc_number)
    directories[shelf_n].append(doc_number)
    print('Документ перемещен.')        
                 
def add_doc():
    number = input('Введите номер документа: ')
    if get_pos_for_doc(number):
        print('Ошибка! Документ с указанным номером уже существует.')
        return
    doc_type = input('Введите тип документа: ')    
    name = input('Введите имя владельца документа: ')    
    shelf = get_shelf()            
    documents.append({'type': doc_type, 'number': number, 'name': name})
    directories[shelf].append(number)
    print('Документ добавлен.')
    return
        
def add_shelf(n=None):
    while True: 
        if n:
            shelf_n = n
        else:
            shelf_n = input('Введите номер новой полки: ')
        try:
            int(shelf_n)
        except:
            print('Ошибка! Номер полки должен быть целым числом.')
            continue
        else: 
            break
    if shelf_n in directories:
        print(f'Полка уже существует!')
    else:        
        directories.setdefault(shelf_n, [])
        print('Полка добавлена.')
    return shelf_n
   
def get_all_docs():
    for doc in documents:
        values_list = list(doc.values())
        #  passport "2207 876234" "Василий Гупкин"
        print(f'{values_list[0]} "{values_list[1]}" "{values_list[2]}"')
        
def del_doc():
    doc_number = input('Введите номер документа: ')
    pos = get_pos_for_doc(doc_number)
    if pos:
        del documents[pos]
        print('Удален из документов.')
    else: return
    shelf = get_shelf_for_doc(doc_number)
    if shelf:
        directories[shelf].remove(doc_number)
        print('Удален с полки.')
    else: return 
       
        
while True:   
    cmd = input('\nВведите команду (h - список доступных команд):\n')
    if cmd == 'h':
        for cmd, dfn in cmds.items():
            print(cmd, ' - ', dfn) 
    elif cmd == 'p': 
        pos = get_pos_for_doc()
        if pos: print(documents[pos]['name'])
    elif cmd == 's':
        shelf = get_shelf_for_doc()
        if shelf: print(shelf)
    elif cmd == 'l': get_all_docs()
    elif cmd == 'as': add_shelf()
    elif cmd == 'd': del_doc()
    elif cmd == 'm': move_doc()        
    elif cmd == 'a': add_doc()        
    elif cmd == 'q': break
    else: print('Неизвестная команда!')
```
